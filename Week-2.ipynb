```python
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
model = LinearRegression()
model.fit(df["a"].values.reshape(-1, 1), df["b"].values.reshape(-1, 1))
```

```python
model.coef_
```

```python
model.intercept_
```

A linear equation with n dimensions can be represented by the equation "y = a0 + a1x1 + a2x2 + a3x3 + .... anxn"

After we fit our training data to Linear Regression model of scikit-learn, coef_ variable of the trained model gives the weights of the unknowns in out equation. intercept_ variable represents a0 in the equation.

In the above case the equation model found is: "y=x+3"

```python
data = [[1, 1], [2, 2], [1, 2], [2, 1], [7, 8], [8, 7], [8, 8], [7, 7]]
Y = [0, 0, 0, 0, 1, 1, 1, 1]
model = LogisticRegression()
model.fit(data, Y)
```

```python
model.classes_
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[0, 0], [9, 9]])
```

```python
model.predict_proba([[0, 0], [9, 9]])
```

Logistic Regression is a classification algorithm. It clusters groups in a supervised fashion.

After training the model predict method gives the class prediction of the sample based on 0.5 threshold directly.
To take the probabilities one needs to use predict_proba. It gives the probabilities of a sample belonging to each class.
